```python
#IQR for SalePrice 
#we Already know the quartile values from the describe
#IQR=Q3-Q1
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data.apply(lambda x:sum(x.isnull()),axis=0)
```

```python
data1 = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
print(data.shape)
data1.shape
```

```python
X=data1.iloc[:,[1,6,7]].values 
y=data1.iloc[:,[8]].values 
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=0) 
```

```python
X_train
```

```python
from sklearn.linear_model.logistic import LogisticRegression
lgsr=LogisticRegression(random_state=0)
lgsr.fit(X_train,y_train) 
```

```python
y_pred=lgsr.predict(X_test) 
y_pred 
```

```python
from sklearn.metrics import confusion_matrix 
conf_matrix=confusion_matrix(y_pred,y_test)
conf_matrix 
```

```python
from sklearn.metrics import accuracy_score 
acc_sco=accuracy_score(y_test,y_pred) 
acc_sco 
```
